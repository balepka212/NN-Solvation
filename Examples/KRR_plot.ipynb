```python
from config import project_path
import pickle as pkl
import numpy as np
from Preprocess.plot_KRR import plot_KRR_blank, plot_KRR
create_manually = False
```

```python
if create_manually:
    with open(project_path('Examples/data/best_KRR2.pkl'), 'rb') as f:
        best_KRR = pkl.load(f)
    best_KRR_MSE = best_KRR

    best_KRR_RMS = {}
    for exp, values in best_KRR_MSE.items():
        best_KRR_RMS[exp] = {}
        for name, v in values.items():
            try:
                best_KRR_RMS[exp][name] = float(v)**(0.5)*2.8032
            except ValueError:
                best_KRR_RMS[exp][name] = v

    with open(project_path('Other/best_KRR_RMS.pkl'), 'wb') as f:
         pkl.dump(best_KRR_RMS, f)

    with open(project_path('Other/best_KRR_RMS.tsv'), 'w') as f:
        headers = best_KRR_RMS['Macro_Blank_KRR1']
        f.write('\t'.join(headers.keys())+'\n')
        for exp, values in best_KRR_RMS.items():
            f.write('\t'.join([str(x) for x in values.values()])+'\n')
```
